```python
import pandas as pd

meeting_file = "./data/싼기타_비싼기타_final.csv"

df = pd.read_csv(meeting_file, encoding="utf-8", sep="|")
display(df)
```

```python
import numpy as np

df["Speaker_Name"] = np.where(df["speaker_id"] == "SPEAKER_01", "김기타", "박기타")
display(df)
```

```python
json_str = df[["Speaker_Name", "text"]].to_json(orient="records", force_ascii=False)
print(json_str)
```

```python
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv()
client = OpenAI(api_key=os.getenv("OPEN_AI_API_KEY"))

response = client.chat.completions.create(
    model="gpt-5-mini",
    messages=[
        {
            "role": "system",
            "content": f"""
            당신은 뛰어난 회의록 요약 및 교정 전문가입니다.
            회의록을 읽고, 주용 내용을 요약하고, 문법 및 맞춤법을 교정합니다.

            출력 형식은 아래 예시를 엄격히 따릅니다.

            # 회의제목
            ## 주요 내용
            ## 참석자별 입장
            ## 결정 사항
            ============================== 이하 회의록 원본 ==============================
            {json_str}
            """,
        },
    ],
)
```

```python
summary = response.choices[0].message.content
print(summary)
```

```python
df.dropna(subset=["text"], inplace=True)
```

```python
from docx import Document

doc = Document() # 새 워드 문서 생성
doc.add_heading('회의 요약 및 교정본', level=1)  # 제목 추가

table = doc.add_table(rows=1, cols=2)  # 표 추가
table.style = 'Table Grid'  # 표 스타일 설정

hdr_cells = table.rows[0].cells # 첫번째 행에 셀 추가
hdr_cells[0].text = 'Speaker'
hdr_cells[1].text = 'Text'

for row in df.itertuples():  # 데이터프레임의 각 행을 순회
    row_cells = table.add_row().cells
    row_cells[0].text = row.Speaker_Name
    row_cells[1].text = row.text

    print(row.Speaker_Name, row.text)

doc.save('./data/회의_요약_및_교정본.docx')  # 워드 문서 저장
```
